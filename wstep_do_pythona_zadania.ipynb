ADANIE 2: Słowniki - Analiza portfolio
---------------------------------------
Stwórz słownik opisujący portfolio inwestycyjne:
```
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}
```

a) Oblicz całkowitą wartość początkową portfolio

b) Oblicz aktualną wartość portfolio

c) Oblicz zysk/stratę dla każdej akcji

d) Które akcje przynoszą największy zwrot (w %)?


````python
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}

wart_pocz = 0
wart_akt = 0
profit = 0

for klucz, wartosc in portfolio.items():
  wart_pocz += (wartosc['liczba']*wartosc['cena_zakupu'])
  wart_akt += (wartosc['liczba']*wartosc['cena_aktualna'])

  profit = (wartosc['cena_aktualna']-wartosc['cena_zakupu'])*wartosc['liczba']
  print(f"{klucz} - {"Zysk" if profit > 0 else "Strata"}: {profit}")

  print(f"Zwrot na akcję w procentach: {(wartosc['cena_aktualna']-wartosc['cena_zakupu'])/wartosc['cena_zakupu']*100}%")

print(f"Portfolio:\nWartość początkowa: {wart_pocz}\nWartość aktualna: {wart_akt}")


````

temp_2022 = [2, 4, 8, 13, 18, 21, 24, 23, 19, 14, 8, 3]

````python
import matplotlib.pyplot as plt

temp_2022 = [2, 4, 8, 13, 18, 21, 24, 23, 19, 14, 8, 3]
temp_2023 = [3, 5, 9, 14, 19, 22, 25, 24, 20, 15, 9, 4]
miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze', 'Lip', 'Sie', 'Wrz', 'Paź', 'Lis', 'Gru']

plt.plot(miesiace, temp_2022, label="2022")
plt.plot(miesiace, temp_2023, label="2023")

plt.title("Zmiana temperatur między latami")
plt.xlabel("Miesiące")
plt.ylabel("Temperatura")
plt.legend()

roznica_miesiecy = []

najzmc23 = 0
najzimniej23 = temp_2023[najzmc23]
najzmc22 = 0
najzimniej22 = temp_2022[najzmc22]

najcmc23 = 0
najcieplej23 = temp_2023[najcmc23]
najcmc22 = 0
najcieplej22 = temp_2022[najcmc22]


for i in range(len(temp_2022)):
  if(temp_2023[i] < najzimniej23):
    nazimniej23 = temp_2023[i]
    najzmc23 = i
  if(temp_2022[i] < najzimniej22):
    nazimniej22 = temp_2022[i]
    najzmc22 = i
  if(temp_2023[i] > najcieplej23):
    najcieplej23 = temp_2023[i]
    najcmc23 = i
  if(temp_2022[i] > najcieplej22):
    najcieplej22 = temp_2022[i]
    najcmc22 = i
  roznica_miesiecy.append(temp_2023[i]-temp_2022[i])

plt.scatter(miesiace[najzmc22], najzimniej22, color="blue", label="Najzimniejszy 2022")
plt.scatter(miesiace[najzmc23], najzimniej23, color="blue", label="Najzimniejszy 2023")

plt.scatter(miesiace[najcmc22], najcieplej22, color="orange", label="Najcieplejszy 2022")
plt.scatter(miesiace[najcmc23], najcieplej23, color="orange", label="Najcieplejszy 2023")

print(f"Średnia różnica temperatur między latami: {sum(roznica_miesiecy)/len(roznica_miesiecy)}")
````

ZADANIE 4: Symulacja Monte Carlo
---------------------------------
Zasymuluj 1000 rzutów kostką (6-ścienną):

a) Wygeneruj dane używając np.random.randint(1, 7, 1000)

b) Stwórz histogram pokazujący częstość każdej liczby

c) Oblicz teoretyczne i empiryczne prawdopodobieństwo każdej liczby

d) Czy wyniki są zgodne z teorią? (każda liczba ~16.67%)

````python
import numpy as np

rzuty = []

for i in range (0,1000):
  rzuty.append(np.random.randint(1,7))

plt.figure(figsize=(10,6))
plt.hist(rzuty, bins=6, edgecolor='black')

prawdopobienstwo = 1/6
w1=0
w2=0
w3=0
w4=0
w5=0
w6 = 0

for wynik in rzuty:
  if wynik == 1:
    w1+=1
  if wynik == 2:
    w2+=1
  if wynik == 3:
    w3+=1
  if wynik == 4:
    w4+=1
  if wynik == 5:
    w5+=1
  if wynik == 6:
    w6+=1
pr1 = w1/len(rzuty)
pr2 = w2/len(rzuty)
pr3 = w3/len(rzuty)
pr4 = w4/len(rzuty)
pr5 = w5/len(rzuty)
pr6 = w6/len(rzuty)

print(f"Prawdopodobieństwo 1 z 6 wynosi: {prawdopobienstwo}\n1: {pr1}\n2: {pr2}\n3: {pr3}\n4: {pr4}\n5: {pr5}\n6: {pr6}")
````
